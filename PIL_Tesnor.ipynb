```python
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load image using PIL
pil_image = Image.open('Face/Face1.jpg')

# Load image using OpenCV
cv2_image = cv2.imread('Face/Face2.jpg')

# Convert OpenCV image to RGB (to match PIL format)
cv2_image_rgb = cv2.cvtColor(cv2_image, cv2.COLOR_BGR2RGB)

# Display both images side by side for comparison
plt.figure(figsize=(12, 6))

# Display PIL image
plt.subplot(1, 2, 1)
plt.imshow(pil_image)
plt.title("PIL Image (RGB)")
plt.axis('off')

# Display OpenCV image (converted to RGB)
plt.subplot(1, 2, 2)
plt.imshow(cv2_image_rgb)
plt.title("OpenCV Image (BGR converted to RGB)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from PIL import Image
import torchvision.transforms as transforms
import torch

# Convert the image to a PyTorch tensor
transform = transforms.ToTensor()  # Converts to a tensor and scales to [0, 1]
tensor_image = transform(pil_image)

# Check the shape of the tensor
print(f"Tensor Shape: {tensor_image.shape}")  # (Channels, Height, Width)

# Get min and max pixel values in the tensor
min_value = torch.min(tensor_image)
max_value = torch.max(tensor_image)

print(f"Tensor Image - Min Pixel Value: {min_value}, Max Pixel Value: {max_value}")
```

```python
import cv2
import torch
import torchvision.transforms as transforms

# Convert BGR to RGB (PyTorch expects RGB images)
cv2_image_rgb = cv2.cvtColor(cv2_image, cv2.COLOR_BGR2RGB)

# Convert the image to a PyTorch tensor
transform = transforms.ToTensor()  # Converts to tensor and scales to [0, 1]
tensor_image = transform(cv2_image_rgb)

# Check tensor properties
print(f"Tensor Shape: {tensor_image.shape}")  # (Channels, Height, Width)
print(f"Min Value: {torch.min(tensor_image)}, Max Value: {torch.max(tensor_image)}")
```

```python
# Convert BGR to RGB (PyTorch expects RGB images)
cv2_image_rgb = cv2.cvtColor(cv2_image, cv2.COLOR_BGR2RGB)

# Convert the OpenCV image to a PyTorch tensor
tensor_image = torch.from_numpy(cv2_image_rgb).permute(2, 0, 1).float()  # (Channels, Height, Width)

# Check tensor properties
print(f"Tensor Shape: {tensor_image.shape}")  # (Channels, Height, Width)
print(f"Min Value: {torch.min(tensor_image)}, Max Value: {torch.max(tensor_image)}")
```
